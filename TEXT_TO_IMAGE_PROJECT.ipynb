# **STEP 2 : LOAD MODEL**

```python
import torch
from diffusers import DiffusionPipeline

# switch to "mps" for apple devices
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, device_map="balanced")

prompt = "Astronaut in a jungle, cold color palette, muted colors, detailed, 8k"
image = pipe(prompt).images[0]
```

# **STEP 3 : CODE INTERFACE**

```python
import torch
from diffusers import StableDiffusionXLPipeline
from PIL import Image
from google.colab import userdata

# Load token from Colab Secrets
HF_TOKEN = userdata.get("HF_TOKEN")
if HF_TOKEN is None:
    raise ValueError("HF_TOKEN not found in Colab Secrets")

# Load SDXL model
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    token=HF_TOKEN
)

# Move to GPU if available and enable CPU offload to save VRAM
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)
pipe.enable_model_cpu_offload()

# Prompt
prompt = "generate iron man on indian road, 8k, cinematic, ultra realistic"

# Generate image
image = pipe(
    prompt=prompt,
    num_inference_steps=30,
    guidance_scale=7.5
).images[0]

# Save image
image.save("output.png")
print("✅ Image saved as output.png")
```

# **STEP 4 : GARDIO INTERFACE**

```python
import gradio as gr

def generate_from_ui(prompt):
    image = pipe(
        prompt,
        num_inference_steps=30,
        guidance_scale=7.5
    ).images[0]
    return image

with gr.Blocks() as demo:
    gr.Markdown("## 🎨 SDXL Text to Image Generator")
    gr.Markdown("Type prompt and click Generate")

    prompt = gr.Textbox(
        label="Prompt",
        placeholder="A cyberpunk city in india, ultra realistic, 8k"
    )

    btn = gr.Button("🚀 Generate Image")
    output = gr.Image(label="Generated Image")

    btn.click(fn=generate_from_ui, inputs=prompt, outputs=output)

demo.launch(share=True)
```
